```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
import cmocean
from eofs.xarray import Eof
```

Open and look at the data:

```python
ds = xr.open_dataset('~/downloads/test2.nc')
ds
```

```python
standardised = (ds['value']-ds['value'].mean(dim='time'))/ds['value'].std(dim='time')
standardised
```

## (1) Hovmoller plot

```python
days = np.array(ds.time)/(24*3600*1e9)
space = np.array(ds.space)
values = np.array(ds.value)

plt.figure(figsize = (6,12))
plt.contourf(space,days,values, cmap = cmocean.cm.balance, levels = 21)
plt.xlabel('Space')
plt.ylabel('Time [days]')
plt.colorbar()
plt.title('Hovmoller plot')
```

Looks like there's some rightward (maybe leftward too but these are fainter) travelling waves with a period ~20 days in there so this should appear in the EOF analysis

## (2) Eigenvalue spectrum with North uncertainty bars

Use the EOFs package to find the EOFs, associated e-values/PCs, and do the North test (assuming each sample independent) directly from the xarray DataArray containing the standardised (in time) data:

```python
solver = Eof(standardised)
eofs = solver.eofs(neofs=None)
pcs = solver.pcs(npcs=None, pcscaling=1)
evalues = solver.eigenvalues(neigs=None)
variance_fractions = solver.varianceFraction(neigs=None)
errs = solver.northTest(neigs=None,vfscaled=True)
```

Plot the eigenvalue spectrum to see how many meaningful EOFs are there and how much variance they explain:

```python
plt.figure(figsize = (12,8))
plt.errorbar(np.arange(10),variance_fractions[:10],errs[:10], capsize = 5)
plt.ylim([0,None])
plt.ylabel('Fraction of variance explained')
plt.xlabel('EOF number')
```

Applying the North test strictly (i.e. if the error bars overlap, then the two different EOFs aren't each special since a linear combination of them will also explain a lot of variance), then only the first EOF is meaningful.

## (3) Plot the structure of the first 5 EOFs:

```python
eof0 = np.array(eofs.sel(mode=0))
eof1 = np.array(eofs.sel(mode=1))
eof2 = np.array(eofs.sel(mode=2))
eof3 = np.array(eofs.sel(mode=3))
eof4 = np.array(eofs.sel(mode=4))

plt.figure(figsize = (16,8))
plt.plot(eof0, label = '1st EOF')
plt.plot(eof1, label = '2nd EOF')
plt.plot(eof2, label = '3rd EOF')
plt.plot(eof3, label = '4th EOF')
plt.plot(eof4, label = '5th EOF')

plt.legend()
plt.xlabel('Space')
plt.ylabel('e-vector amplitude')
```

## (4) Plot first 4 PC time-series:

```python
pc0 = np.array(pcs.sel(mode=0))
pc1 = np.array(pcs.sel(mode=1))
pc2 = np.array(pcs.sel(mode=2))
pc3 = np.array(pcs.sel(mode=3))

fig, axs = plt.subplots(2,2,figsize = (20,16))
axs[0,0].plot(pc0, label = '1st EOF')
axs[0,1].plot(pc1, label = '2nd EOF')
axs[1,0].plot(pc2, label = '3rd EOF')
axs[1,1].plot(pc3, label = '4th EOF')

axs[0,0].set_xlabel('Time')
axs[0,0].set_ylabel('Normalised PC')
axs[0,0].set_title('1st EOF')

axs[0,1].set_xlabel('Time')
axs[0,1].set_ylabel('Normalised PC')
axs[0,1].set_title('2nd EOF')

axs[1,0].set_xlabel('Time')
axs[1,0].set_ylabel('Normalised PC')
axs[1,0].set_title('3rd EOF')

axs[1,1].set_xlabel('Time')
axs[1,1].set_ylabel('Normalised PC')
axs[1,1].set_title('4th EOF')
```

The first PC series clearly picks out the 20 day oscillations we saw in the Hovmoller, it looks like EOFs 2&3 together are picking out another higher frequency (~10 day) set of travelling waves.

## (5) Regression of first 5 EOFs onto the un-normalised data:

We can now regress the normalised PCs onto the un-scaled (I'll subtract the mean before the regression then add it back in before plotting, might be unnecessary due to linearity...?) original data:

```python
X = np.array(ds['value']-ds['value'].mean(dim = 'time'))
mu = np.array(ds['value'].mean(dim = 'time'))

D0 = np.matmul(X.T,pc0)/120+mu
D1 = np.matmul(X.T,pc1)/120+mu
D2 = np.matmul(X.T,pc2)/120+mu
D3 = np.matmul(X.T,pc3)/120+mu
D4 = np.matmul(X.T,pc4)/120+mu

plt.figure(figsize = (12,8))
plt.plot(D0, label = '1st EOF')
plt.plot(D1, label = '2nd EOF')
plt.plot(D2, label = '3rd EOF')
plt.plot(D3, label = '4th EOF')
plt.plot(D4, label = '5th EOF')
plt.legend()
plt.xlabel('Space')
plt.ylabel('Value')
plt.title('EOFs from regressing PCs onto data');
```

## (6) Discuss

Based on this analysis I would propose to further study the structures captured by the first 3 EOFs. 

Concretely, the first EOF captures a wave with a time period of ~20 days which is the most immediately prominent feature visually in the Hovmoller plot and this EOF explains the most variance in the observations. Since there is only one mode with this wavelength, this longer period variability looks to be associated with a standing wave. This is consistent with the appearance of both leftward and rightward travelling waves in the Hovmoller, these combine to produce a standing wave.

The second and third EOFs' structures cannot really be separated from one another based on the North test error bars and their structure looks to be sinusoidal with the same wavelength but a phase shift. This implies that together they could form the basis of a travelling wave (which requires both a wavelength and a phase shift) with time period of ~10 days. This is consistent with the appearance of some smaller-scale periodic structures in the Hovmoller.

While the fourth and fifth EOFs also look like shifted sinusoids of the same wavelength (longer wavelength than the 2nd-3rd couplet), the PC time series' show no obvious temporal periodicity. Given that the wavelengths look to be roughly the same as the domain size, I'm interpreting these EOFs to merely be the beginning of a Fourier series expansion for the analysis domain, rather than any physically meaningful phenomenon. They do seem to explain significantly more variance than the 6th and onwards e-values (this looks to be where the continuum begins) but I cannot discern a clear enough temporal structure in the PC time series to label them as physically meaningful... my take on EOFs 4 and 5 is that they aren't physical given this data, but I could believe that they might be the faint signal of something physical which could be clearer with more data.
