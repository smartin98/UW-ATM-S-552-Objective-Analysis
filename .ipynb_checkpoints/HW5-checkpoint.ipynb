```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
import cmocean
from scipy import stats
from scipy.signal import detrend
import cartopy.crs as ccrs
from eofs.xarray import Eof
```

First let's consider the global oceans between 65S and 65N:

```python
ds = xr.open_dataset('~/downloads/sst.mnmean.nc')
ds = ds.sel(time=slice('1960','2018')).sel(lat=slice(65,-65))
ds
```

```python
clm = ds.sst.groupby('time.month').mean(dim='time')
```

```python
clm.sel(month=12).plot()
```

```python
anom = ds.sst.groupby('time.month') - clm
weights = np.cos(np.deg2rad(ds.lat)).where(~anom[0].isnull())
weights /= weights.mean()
```

```python
weights.plot()
```

```python
anom.isel(time=-12).plot()
```

```python
anom_detrended = xr.apply_ufunc(detrend, anom.fillna(0),
                                    kwargs={'axis': 0}).where(~anom.isnull())
```

```python
anom_detrended.isel(time=-12).plot()
```

Use this handy python package to calculate the EOFs, PCs, variance fractions, etc.

```python
solver = Eof(anom_detrended, weights=np.sqrt(weights))
eofs = solver.eofs(neofs=None)
pcs = solver.pcs(npcs=None, pcscaling=1)
evalues = solver.eigenvalues(neigs=None)
variance_fractions = solver.varianceFraction(neigs=None)
errs = solver.northTest(neigs=None,vfscaled=True)
```

Fraction of variance explained by each EOF with error bar estimated using North test (assuming every month is an independent sample)

```python
plt.figure(figsize = (12,8))
plt.errorbar(np.arange(10),variance_fractions[:10],errs[:10], capsize = 5)
plt.ylim([0,None])
plt.ylabel('Fraction of variance explained')
plt.xlabel('EOF number')
```

The PCs were normalised to have unit variance, the EOFs are un-normalised since the data weren't standardised, so scale the PCs by the rms of the respective EOFs to get physical amplitudes for PC time-series'. Based on the above spectrum I choose to truncate the EOFs beyond the 4th EOF.

```python
fig, axs = plt.subplots(2,2,figsize = (16,16))

e0 = np.array(eofs.sel(mode=0))
e0 = e0[~np.isnan(e0)]
rms0 = np.nanmean(e0**2)**0.5

e1 = np.array(eofs.sel(mode=1))
e1 = e1[~np.isnan(e1)]
rms1 = np.nanmean(e1**2)**0.5

e2 = np.array(eofs.sel(mode=2))
e2 = e2[~np.isnan(e2)]
rms2 = np.nanmean(e2**2)**0.5

e3 = np.array(eofs.sel(mode=3))
e3 = e3[~np.isnan(e3)]
rms3 = np.nanmean(e3**2)**0.5

(pcs.sel(mode=0)*rms0).plot(ax = axs[0,0])
(pcs.sel(mode=1)*rms1).plot(ax = axs[0,1])
(pcs.sel(mode=2)*rms2).plot(ax = axs[1,0])
(pcs.sel(mode=3)*rms3).plot(ax = axs[1,1])
```

Regression maps for first 4 EOFs:

```python
fig, axs = plt.subplots(2,2,figsize = (16,16), constrained_layout = True)
fig.suptitle('Regression maps for each EOF [$^{\circ}C$]')
(eofs.sel(mode=0)).plot(ax = axs[0,0])
(eofs.sel(mode=1)).plot(ax = axs[0,1])
(eofs.sel(mode=2)).plot(ax = axs[1,0])
(eofs.sel(mode=3)).plot(ax = axs[1,1])
```

The first mode is dominated by Pacific variability with a spatial pattern that is reminiscent of ENSO, the PC time-series also looks to have a dominant period of variability of around 7 years, consistent with it being a proxy for ENSO. There looks to be an ENSO signal projecting onto all the other 3 EOFS too, evidenced both by the spatial patterns in the Pacific and the ~7 year periodicity in the PC time-series'. A notable feature in the North Atlantic appears in both the second and third EOFs, consistent with the AMO having a projection onto these EOFs. This is further evidenced by the appearance of decadal variability in these PC time-series'

Use this handy python package to calculate the EOFs, PCs, variance fractions, etc.

```python
fig, axs = plt.subplots(3,2,figsize = (16,20))

e0 = np.array(eofs.sel(mode=0))
e0 = e0[~np.isnan(e0)]
rms0 = np.nanmean(e0**2)**0.5

e1 = np.array(eofs.sel(mode=1))
e1 = e1[~np.isnan(e1)]
rms1 = np.nanmean(e1**2)**0.5

e2 = np.array(eofs.sel(mode=2))
e2 = e2[~np.isnan(e2)]
rms2 = np.nanmean(e2**2)**0.5

e3 = np.array(eofs.sel(mode=3))
e3 = e3[~np.isnan(e3)]
rms3 = np.nanmean(e3**2)**0.5

e4 = np.array(eofs.sel(mode=4))
e4 = e4[~np.isnan(e4)]
rms4 = np.nanmean(e4**2)**0.5

(pcs.sel(mode=0)*rms0).plot(ax = axs[0,0])
(pcs.sel(mode=1)*rms1).plot(ax = axs[0,1])
(pcs.sel(mode=2)*rms2).plot(ax = axs[1,0])
(pcs.sel(mode=3)*rms3).plot(ax = axs[1,1])
(pcs.sel(mode=4)*rms4).plot(ax = axs[2,0])
```

Regression maps for first 4 EOFs:

```python
fig, axs = plt.subplots(3,2,figsize = (16,16), constrained_layout = True)
fig.suptitle('Regression maps for each EOF [$^{\circ}C$]')
(eofs.sel(mode=0)).plot(ax = axs[0,0])
(eofs.sel(mode=1)).plot(ax = axs[0,1])
(eofs.sel(mode=2)).plot(ax = axs[1,0])
(eofs.sel(mode=3)).plot(ax = axs[1,1])
(eofs.sel(mode=4)).plot(ax = axs[2,0])
```

The first mode is highly suggestive of the PDO due to its spatial structure and the decadal variability evident in the PC time series. The second and fourth EOFs look like they may have a projection of the AMO based on the PC time series. The third EOD seems to also be dominated by the PDO. Despite explaining a decent fraction of the variance, the 5th EOF PC time series doesn't seem to have a coherent interannual pattern, the spatial structure looks like this could just be a Fourier mode. All the EOFs look to have an ENSO projection in the PC time series.
