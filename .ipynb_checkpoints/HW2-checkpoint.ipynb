```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy.optimize import curve_fit
from scipy.stats import kurtosis
from scipy.stats import skew
from scipy.stats import mode
from tabulate import tabulate
import pandas as pd
```

```python
# Import text file containing test data
x = np.loadtxt('Data_HW_2.txt')   
length = x.shape[0]
x0 = x[:,0]
x1 = x[:,1]

t = np.linspace(0,length,length)   # Make a linear series to be like the time
```

```python
plt.figure(figsize=(16,8))
plt.plot(t,x0, label = 'x0')
plt.plot(t,x1,alpha = 0.5, label = 'x1')
plt.hlines(np.mean(x0),t[0],t[-1], color='red', label = 'x0 mean')
plt.hlines(np.mean(x1),t[0],t[-1], color='green', label = 'x1 mean')
plt.hlines(np.median(x0),t[0],t[-1], color='red', linestyle='--', label = 'x0 median')
plt.hlines(np.median(x1),t[0],t[-1], color='green', linestyle='--', label = 'x1 median')

plt.legend();
```

## Part 1

t-test to test if the means and medians are different (H0=means same, H1=means different)

```python
tstat = stats.ttest_ind(x0,x1,equal_var=False) 
print(tstat)
```

So t-test rejects null hypothesis with p=0.0001, i.e. means are likely different

Now do rakn sum test instead:

```python
rstat = stats.ranksums(x0,x1)
print(rstat)
```

Confidence level from this test much lower than from t-test (only p=0.2)

## Part 2
### (a)
Find empirical distributions

```python
nbins = 20 
counts0, bins0 = np.histogram(x0,nbins)
counts1, bins1 = np.histogram(x1,nbins)
bins0 = (bins0[:-1] + bins0[1:])/2
bins1 = (bins1[:-1] + bins1[1:])/2
plt.bar(bins0,counts0, label = 'x0')
plt.bar(bins1,counts1, alpha=0.5, label = 'x1')
plt.legend();
```

### (b) & (c)
Fit empirical distributions using scipy.optimize.curve_fit

```python
# Define the Gaussian function
def Gauss(x, A, B, C):
    y = A*np.exp(-1*B*(x-C)**2)
    return y
def Gamma(x, A, B, C):
    y=C*x**(A-1)*np.exp(-x/B)
    return y
parameters0_gaussian, covariance0_gaussian = curve_fit(Gauss, bins0, counts0)
parameters0_gamma, covariance0_gamma = curve_fit(Gamma, bins0, counts0) 
A0_gauss = parameters0_gaussian[0]
B0_gauss = parameters0_gaussian[1]
C0_gauss = parameters0_gaussian[2]
A0_gamma = parameters0_gamma[0]
B0_gamma = parameters0_gamma[1]
C0_gamma = parameters0_gamma[2]
  
fit_x0_gauss = Gauss(bins0, A0_gauss, B0_gauss, C0_gauss)
fit_x0_gamma = Gamma(bins0, A0_gamma, B0_gamma, C0_gamma)

parameters1_gaussian, covariance1_gaussian = curve_fit(Gauss, bins1, counts1)
parameters1_gamma, covariance1_gamma = curve_fit(Gamma, bins1, counts1) 
A1_gauss = parameters1_gaussian[0]
B1_gauss = parameters1_gaussian[1]
C1_gauss = parameters1_gaussian[2]
A1_gamma = parameters1_gamma[0]
B1_gamma = parameters1_gamma[1]
C1_gamma = parameters1_gamma[2]
  
fit_x1_gauss = Gauss(bins1, A1_gauss, B1_gauss, C1_gauss)
fit_x1_gamma = Gamma(bins1, A1_gamma, B1_gamma, C1_gamma)


fig, axs = plt.subplots(1,2,figsize=(16,8))
axs[0].bar(bins0, counts0, label = 'Data')
axs[0].plot(bins0, fit_x0_gauss, label = 'Gaussian', color = 'orange')
axs[0].plot(bins0, fit_x0_gamma, label = 'Gamma', color = 'red')
axs[0].set_title('x0')
axs[1].bar(bins1, counts1, label = 'Data')
axs[1].plot(bins1, fit_x1_gauss, label = 'Gaussian', color = 'orange')
axs[1].plot(bins1, fit_x1_gamma, label = 'Gamma', color = 'red')
axs[1].set_title('x1')
axs[0].legend()
axs[1].legend()
```

### (d)
table with stats for each variable:

```python
mode0, count = mode(x0)
mode1, count = mode(x1)
mode0 = mode0[0]
mode1 = mode1[0]
mean0 = np.mean(x0)
mean1 = np.mean(x1)
std0 = np.std(x0)
std1 = np.std(x1)
skew0 = skew(x0)
skew1 = skew(x1)
kurt0 = kurtosis(x0)
kurt1 = kurtosis(x1)
data = [['x0', mean0, mode0, std0, skew0, kurt0],['x1', mean1, mode1, std1, skew1, kurt1]]
df = pd.DataFrame(data)
print(tabulate(df, showindex=False, headers=['Dataset','Mean','Mode','STD','Skewness','Kurtosis'], floatfmt=".2f"))
```

### (e)/(f)
See part 1 and below:

The two tests give very different answers: t-test (which tests the difference in the means of the datasets) gives a a high significance level (p=0.001) for rejecting the null hypothesis. Whereas the the Wilcoxon rank sum test (which tests the difference in the medians) gives a much lower significance level (p=0.2). This is since x1 has a high skewness, so the mean and median will be very different, so the two tests will give very different results.
